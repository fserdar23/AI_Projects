```python
# 2 sütunu birlestirdik ve tek bir sutun olusturduk
```

```python
df["birlesik"]=df['premise'] + ' ' + df['hypothesis']
```

```python
df
```

```python
df["birlesik"].iloc[5]
```

```python
from sklearn.pipeline import Pipeline

from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
x=df["birlesik"]
y=df["label"]
```

```python
clf=Pipeline([("tfidf",TfidfVectorizer()),("classifier",GradientBoostingClassifier())])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.25)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dsc=Pipeline([("tfidf",TfidfVectorizer()),("classifier",DecisionTreeClassifier())])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.25)
```

```python
dsc.fit(x_train,y_train)
```

```python
y_pred=dsc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
```
