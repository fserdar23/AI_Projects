### Web Scraping

```python
import pandas as pd

pd.set_option("display.max_columns",100)

import seaborn as sns
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup
import requests
```

```python
basliklar=[]
fiyatlar=[]
banyo_sayilari=[]
metrekare=[]
Oda_sayisi=[]
kat=[]
isitma=[]
yas=[]
tip=[]
semt=[]

for i in range(1, 50):  # 1'den 50'ye kadar döngü
    url = f"https://www.emlakjet.com/kiralik-konut/istanbul/{i}/"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    
    # İlanları bul
    linkler=soup.find_all("div",{"class":"_3qUI9q"}) 
      
    for link in linkler:
        link1="https://www.emlakjet.com"+link.a.get("href")
        page2=requests.get(link1)
        soup2=BeautifulSoup(page2.content,"html.parser")
        #Fiyat
        div_fiyat = soup2.select_one("div.OqBSwC[data-scroll='fiyat-analizi']")
        fiyat = div_fiyat.find_next_sibling(text=True).strip()
        # Baslik
        baslik =soup2.h1.text
        
        p_element = soup2.select_one("div._3VQ1JB p")

        # p öğesinin içeriğini alma
        if p_element:
            ilce = p_element.get_text()       
       
        divler = soup2.find_all("div", {"class":"_1bVOdb"})  # bilgilerini içeren div'leri bul
    
        banyo_sayisi = None
        metre = None
        oda = None
        kats = None
        ist = None
        ys = None
        tipi = None
        
        for i in range(0, len(divler), 2):  # İki adımda ilerle
            if "Banyo Sayısı" in divler[i].text:  # Başlık div'ini kontrol et
                banyo_sayisi = divler[i + 1].text.strip()  # Sonraki div'den sayıyı al
                
            elif "Net Metrekare" in divler[i].text: 
                metre = divler[i + 1].text.strip()  
                
            elif "Oda Sayısı" in divler[i].text:  
                oda = divler[i + 1].text.strip()  
                
            elif "Binanın Kat Sayısı" in divler[i].text: 
                kats = divler[i + 1].text.strip()  
        
            elif "Isıtma Tipi" in divler[i].text:  
                ist = divler[i + 1].text.strip()  
                
            elif "Binanın Yaşı" in divler[i].text:
                ys = divler[i + 1].text.strip()  
                
            elif "Tipi" in divler[i].text:  
                tipi = divler[i + 1].text.strip()  
            
        semt.append(ilce)
        basliklar.append(baslik)
        fiyatlar.append(fiyat)
        metrekare.append(metre)
        Oda_sayisi.append(oda)
        banyo_sayilari.append(banyo_sayisi)
        kat.append(kats)
        isitma.append(ist)
        yas.append(ys)
        tip.append(tipi)
```

```python
df=pd.DataFrame(columns=["semt","baslik","fiyat","metrekare","oda_sayisi","banyo_sayisi","kat","isitma","yas","tip"])
```

```python
df["baslik"]=basliklar
df["semt"]=semt
df["fiyat"]=fiyatlar
df["metrekare"]=metrekare
df["oda_sayisi"]=Oda_sayisi
df["banyo_sayisi"]=banyo_sayilari
df["kat"]=kat
df["isitma"]=isitma
df["yas"]=yas
df["tip"]=tip
```

```python
df.to_csv("Ist.House.csv")
```

### Feeature Engineering

```python
df['fiyat'] = df['fiyat'].str.replace(',', '').astype(int)
```

```python
df['metrekare'] = df['metrekare'].str.split().str[0]

df['metrekare'] = df['metrekare'].str.replace('.', '').astype(float)
```

```python
df["oda_sayisi"]=df["oda_sayisi"].str.replace("Stüdyo","1+0")
df["oda_sayisi"]=df["oda_sayisi"].str.replace("1 Oda","1+0")
df["oda_sayisi"]=df["oda_sayisi"].str.replace("4.5","5")
df["oda_sayisi"]=df["oda_sayisi"].str.replace("1.5","2")
df["oda_sayisi"]=df["oda_sayisi"].str.replace("2.5","3")
df["oda_sayisi"]=df["oda_sayisi"].str.replace("3.5","4")
df["oda_sayisi"]=df["oda_sayisi"].str.replace("9+ Oda","10+0")

df["oda_sayisi"]=df["oda_sayisi"].fillna("3+1")

df['oda_sayisi'] = df['oda_sayisi'].apply(lambda x: sum(map(float, x.split('+'))))
```

```python
df['banyo_sayisi'] = df['banyo_sayisi'].str.replace("Yok","0")
df['banyo_sayisi'] = df['banyo_sayisi'].str.replace("6+","7")

df['banyo_sayisi'] = df['banyo_sayisi'].fillna("1")

df['banyo_sayisi'] = df['banyo_sayisi'].astype(float)

df['banyo_sayisi'] = df['banyo_sayisi']
```

```python
df['kat'] = df['kat'].fillna("4")

df['kat'] = df['kat'].astype(float)
```

```python
df['yas'] = df['yas'].str.replace("0 (Yeni)","0")
df['yas'] = df['yas'].str.replace("5-10","7")
df['yas'] = df['yas'].str.replace("11-15","13")
df['yas'] = df['yas'].str.replace("16-20","18")
df['yas'] = df['yas'].str.replace("21 Ve Üzeri","25")

df['yas'] = df['yas'].fillna("7")

df['yas'] = df['yas'].astype(float)
```

```python
df["isitma"]=df["isitma"].fillna("Isıtma Yok")
```

```python
df['tip'] = df['tip'].replace(['Çiftlik Evi', 'Kagir',"Villa","Bina"], 'Müstakil Ev')
df['tip'] = df['tip'].replace(['AÃ§Ä±k Balkon', 'Açık Balkon, Kapalı Teras',"Açık Balkon, Kapalı Balkon",
                               "Bina","Açık Balkon, Açık Teras","Fransız Balkon"], 'Açık Balkon')
df['tip'] = df['tip'].replace(["bina","Çelik"],"Daire")
df["tip"] = df["tip"].replace(["Kapalı Balkon, Kapalı Teras","Kapalı Teras"],"Kapalı Balkon")
```

```python
df['semt'] = df['semt'].str.split(' - ').str[1]
```

### Modelling

```python
df
```

```python
x=df.drop(["Unnamed: 0","fiyat","baslik"],axis=1)
y=df["fiyat"]
```

```python
x=pd.get_dummies(x)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns",100)

from sklearn.linear_model import LinearRegression,SGDRegressor,Ridge,Lasso,ElasticNet
from sklearn.neighbors import KNeighborsRegressor, RadiusNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor, plot_tree, ExtraTreeRegressor
#pip install xgboost
from xgboost import XGBRegressor
from sklearn.svm import SVR

from sklearn.neural_network import MLPRegressor

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error

from sklearn.preprocessing import StandardScaler


def algo_test(x,y):
        #Bütün modelleri tanımlıyorum
        L=LinearRegression()
        R=Ridge()
        Lass=Lasso()
        E=ElasticNet()
        sgd=SGDRegressor()
        ETR=ExtraTreeRegressor()
        GBR=GradientBoostingRegressor()
        kn=KNeighborsRegressor()
        rkn=RadiusNeighborsRegressor(radius=1.0)
        ada=AdaBoostRegressor()
        dt=DecisionTreeRegressor()
        xgb=XGBRegressor()
        svr=SVR()
        mlp_regressor = MLPRegressor()

       
        
        algos=[L,R,Lass,E,sgd,ETR,GBR,ada,kn,dt,xgb,svr,mlp_regressor]
        algo_names=['Linear','Ridge','Lasso','ElasticNet','SGD','Extra Tree','Gradient Boosting',
                    'KNeighborsRegressor','AdaBoost','Decision Tree','XGBRegressor','SVR','mlp_regressor']
        
        x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=.20,random_state=42)
        
        r_squared= []
        rmse= []
        mae= []
        
        #Hata ve doğruluk oranlarını bir tablo haline getirmek için bir dataframe oluşturuyorum
        result=pd.DataFrame(columns=['R_Squared','RMSE','MAE'],index=algo_names)
        
        
        for algo in algos:
            p=algo.fit(x_train,y_train).predict(x_test)
            r_squared.append(r2_score(y_test,p))
            rmse.append(mean_squared_error(y_test,p)**.5)
            mae.append(mean_absolute_error(y_test,p))
        
            

        #result adlı tabloya doğruluk ve hata oranlarımı yerleştiriyorum
        result.R_Squared=r_squared
        result.RMSE=rmse
        result.MAE=mae
        
       #oluşturduğum result tablosunu doğruluk oranına (r2_score) göre sıralayıp dönüyor
        rtable=result.sort_values('R_Squared',ascending=False)
        return rtable
```

```python
algo_test(x,y)
```

### Derin Öğrenme

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale

from sklearn.metrics import r2_score, mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,BatchNormalization,Dropout
```

```python
x=scale(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20 ,random_state=42)
```

```python
model=Sequential()
model.add(Dense(256,activation="relu")) 
model.add(Dense(196,activation="relu")) 
model.add(Dense(144,activation="relu")) 
model.add(BatchNormalization())
model.add(Dropout(.25))
model.add(Dense(120,activation="relu")) 
model.add(Dense(96,activation="relu")) 
model.add(Dense(64,activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(.25))
model.add(Dense(30,activation="relu"))
model.add(Dense(20,activation="relu"))
model.add(Dense(4,activation="relu"))
model.add(Dense(1)) 
model.compile(loss="mse", optimizer="adam")                                                                         
```

```python
history=model.fit(x_train,y_train ,validation_data=(x_test,y_test), batch_size=32, epochs=500, verbose=1)
```

```python
tahmin=model.predict(x_test)
```

```python
mean_squared_error(tahmin,y_test)**.5
```

```python
r2_score(tahmin,y_test)
```

```python
loss_f=pd.DataFrame(history.history)
```

```python
loss_f.plot()
```
