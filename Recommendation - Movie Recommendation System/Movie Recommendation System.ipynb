```python
df=df[["genres","keywords","title","popularity","production_companies","vote_average","vote_count"]]
```

```python
df=df[df["vote_count"]>300]
```

```python
import json

# Kategori İsimleri çıkartma fonksiyonu
def extract_names(json_str):
    try:
        # JSON string'ini listeye dönüştürme
        genres = json.loads(json_str)
        # Her bir nesneden ismi alma
        return [genre['name'] for genre in genres]
    except (json.JSONDecodeError, KeyError):
        return None  # Hata durumunda None döndür

# İsimleri çıkartma
df['genre_names'] = df['genres'].apply(extract_names)
```

```python
df['keywords_name'] = df['keywords'].apply(extract_names)
```

```python
pop=df[["popularity","title","vote_average"]]
```

```python
df = df.explode('genre_names')
```

<b>Kategoriye göre kosinüs benzerliği uyguluyoruz</b>

```python
csns_matrix=df.pivot_table(index='title', columns=df["genre_names"].values, values='genre_names', aggfunc='count', fill_value=0)
```

```python
csns_matrix.head()
```

```python
csns_matrix=pd.merge(csns_matrix,pop,on="title")
```

```python
csns_matrix.head(1)
```

```python
csns_matrix = csns_matrix.reindex(columns=[csns_matrix.columns[-1]] + list(csns_matrix.columns[:-1]))
```

```python
csns_matrix = csns_matrix.reindex(columns=[csns_matrix.columns[-1]] + list(csns_matrix.columns[:-1]))
```

```python
csns_matrix.head(1)
```

```python
movie_dict = {}

for index, row in csns_matrix.iterrows():
    rating = row['vote_average']  # Ürün kimliği
    name = row['title']  # Ürün kimliği
    genres = [row[col] for col in csns_matrix.columns[3:]]  # 4. sütundan sonraki değerler
    popul=row["popularity"]
    movie_dict[name]=(name,rating,popul,genres)
```

```python
movie_dict
```

```python
from scipy import spatial

def compute_distance(a,b):
    genre_distance=spatial.distance.cosine(a[3],b[3])
    popularity_distance=abs(a[2]-b[2])
    return genre_distance+popularity_distance
```

```python
movie_dict["Avatar"]
```

```python
movie_dict["Pirates of the Caribbean: At World's End"]
```

```python
compute_distance(movie_dict["Avatar"],movie_dict["Pirates of the Caribbean: At World's End"]) #2 vektör arasındaki mesafe --- Bu mesafe ne kadar az çıkarsa 2 film o kadar benzer oluyor
```

```python
mesafeler=[]
for movie1 in movie_dict:
    for movie2 in movie_dict:
        mesafeler.append(compute_distance(movie_dict[movie1],movie_dict[movie2]))
```

```python
mesafeler
```

```python
input_movie = input("Bir film adı girin: ")

# Girilen film ile diğer filmler arasındaki mesafeleri hesaplamak
if input_movie in movie_dict:
    input_movie_data = movie_dict[input_movie]
    
    # Mesafeleri hesaplama
    distances = {}
    for movie, data in movie_dict.items():
        if movie != input_movie:
            distances[movie] = compute_distance(input_movie_data, data)
    
    # En benzer 5 filmi bulma
    similar_movies = sorted(distances.items(), key=lambda x: x[1])[:5]
    
    # Sonuçları gösterme
    print("En benzer 5 film:")
    for movie, distance in similar_movies:
        print(f"{movie} - Mesafe: {distance:.2f}")
else:
    print("Girilen film bulunamadı.")
```

```python
with open("movies.json", "w") as f:
    json.dump(movie_dict,f)
```
