### Feature Engineering

```python
yeni["Age"]=yeni["Age"].fillna(yeni["Age"].mean())
```

```python
yeni["Embarked"]=yeni["Embarked"].fillna("S")
```

```python
yeni["Fare"]=yeni["Fare"].fillna(yeni["Fare"].mean())
```

```python
yeni.describe()
```

```python
yeni.corr(numeric_only=True)
```

```python
yeni=yeni.drop("Cabin",axis=1)
```

```python
yeni=yeni.drop("Name",axis=1)
```

```python
yeni=yeni.drop("Ticket",axis=1)
```

```python
yeni.isnull().sum()
```

```python
ttrain=yeni.iloc[:891]
```

```python
ttest=yeni.iloc[891:]
```

```python
ttrain.isnull().sum()
```

```python
ttest.isnull().sum()
```

```python
ttrain= pd.get_dummies(ttrain, drop_first=True)
```

```python
ttrain.head()
```

```python
x=ttrain.drop("Survived",axis=1)
y=ttrain["Survived"]
```

```python
ttest= pd.get_dummies(ttest, drop_first=True)
```

```python
ttest=ttest.drop("Survived",axis=1)
```

```python
ttest.head()
```

```python
x.head()
```

### Modelling

```python
b=BernoulliNB()
```

```python
b.fit(x,y)
```

```python
x.head()
```

```python
ttest.head()
```

```python
ttest["Survived"]=b.predict(ttest)
```

```python
tahmin=ttest
```
