```python
import pandas as pd
from collections import Counter
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer

def faster(data):
    # "text" sütunundaki tüm metinleri tek bir metin haline getirme
    all_text = ' '.join(data['text'])
    
    # Metni tokenize etme (kelimelere ayırma)
    words = all_text.lower().split()
    
    # İngilizce stopword'leri çıkarma
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]
    
    # Sentiment analizi yapma
    sia = SentimentIntensityAnalyzer()
    word_scores = [sia.polarity_scores(word)['compound'] for word in filtered_words]
    
    # Olumlu ve olumsuz kelime gruplarını oluşturma
    positive_words = [word for word, score in zip(filtered_words, word_scores) if score > 0]
    negative_words = [word for word, score in zip(filtered_words, word_scores) if score < 0]
    
    # Olumlu ve olumsuz kelimelerin frekanslarını hesaplama
    positive_counts = Counter(positive_words)
    negative_counts = Counter(negative_words)
    
    # Sonuçları yazdırma
    print("En çok kullanılan olumlu kelimeler:")
    for word, count in positive_counts.most_common(10):
        print(f"{word}: {count}")
    
    print("\nEn çok kullanılan olumsuz kelimeler:")
    for word, count in negative_counts.most_common(10):
        print(f"{word}: {count}")
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

def sent(df):
    # "text" sütunundaki tüm metinleri tek bir metin haline getirme
    all_text = ' '.join(df)
    
    # Metni tokenize etme (kelimelere ayırma)
    words = all_text.lower().split()
    
    # Sentiment analizi yapma
    sia = SentimentIntensityAnalyzer()
    word_scores = [sia.polarity_scores(word)['compound'] for word in words]
    
    # Olumlu ve olumsuz kelimelerin sayısını hesaplama
    positive_count = sum(1 for score in word_scores if score > 0)
    negative_count = sum(1 for score in word_scores if score < 0)
    
    # Toplam kelime sayısını hesaplama
    total_count = len(word_scores)
    
    # Yüzdeleri hesaplama
    positive_percentage = (positive_count / total_count) * 100
    negative_percentage = (negative_count / total_count) * 100
    
    return positive_percentage, negative_percentage
```

```python
sent(depo)
```

```python
def wc(data,bgcolor):
    plt.figure(figsize=(10,10))
    mask=np.array(Image.open("cloud.png"))
    wc=WordCloud(background_color=bgcolor, stopwords=STOPWORDS, mask=mask)
    wc.generate(" ".join(data))  # join Metindeki satırları birleştiri - generate En çok tekrar edenleri saydı
    plt.imshow(wc)
    plt.axis("off")
    #plt.savefig("ccccc.png", dpi=300)
```
